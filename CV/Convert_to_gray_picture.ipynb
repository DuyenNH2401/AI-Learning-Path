```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def compute_average(vector):
    return vector.mean()
```

```python
#convert picture to array
r_img = cv2.imread("raw_pic.jpg", 1)
#resize the picture
img = cv2.resize(r_img, (640, 480))
#calculate mean of RGB
img = np.apply_along_axis(compute_average, axis = 2, arr = img)
```

```python
#show image
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(r_img)
axes[0].set_title('Raw picture')
axes[0].axis('off')

axes[1].imshow(img, cmap='gray')
axes[1].set_title('Gray picture')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
