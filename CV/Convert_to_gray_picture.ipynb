```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def compute_average(vector):
    return vector.mean()
```

```python
#convert picture to array
img = cv2.imread("image3.jpg", 1)
#resize the picture
img = cv2.resize(img, (640, 480))
#calculate mean of RGB
img = np.apply_along_axis(compute_average, axis = 2, arr = img)

#show
img = img.astype(np.uint8)
plt.imshow(img, cmap="gray")
plt.axis('off')
plt.show()
```
